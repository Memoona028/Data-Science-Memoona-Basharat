```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats 
```

- date_range() func used to create Datatime index from start to end date 
- .series() func we used here too cerate a random series of sales w.r.t to Datetime index
- and then we plotted it using .plot() func

```python
start_date = "Jan 1, 2024"
end_date = "July 28, 2024"
idx = pd.date_range(start_date,end_date)
print(f"Length of index is {len(idx)}")
#len(idx) length of date range we set above
items_sold = pd.Series(np.random.randint(25,50,size=len(idx)),index=idx)
items_sold.plot(figsize=(10,6))
```

```python

index = [1, 2, 3, 4, 5]
name = ["Memoona", "Areeba", "Nimra", "Moneeba", "Hina"]
name_series = pd.Series(name, index=index)
print(name_series)
```

```python
additional_items = pd.Series(10,index=pd.date_range(start_date,end_date,freq="W"))
# here  we will add two series together to get total number:

total_items = items_sold.add(additional_items,fill_value=0)
total_items.plot()
```

```python
monthly = total_items.resample("1M").mean()
ax = monthly.plot(kind='bar')
```

```python
# Dimensionality Reduction
from sklearn.decomposition import PCA
numeric_cols = ['numcol', 'yieldpercol', 'totalprod', 'stocks', 'priceperlb', 'prodvalue', 'year']

# Now applying PCA on numeric cols
pca = PCA(n_components=2)
df_pca = pca.fit_transform(df[numeric_cols])

# Creating a DataFrame with the PCA results
df_pca = pd.DataFrame(df_pca, columns=['PCA1', 'PCA2'])

print(df_pca)
```
